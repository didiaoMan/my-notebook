血压app顾问端功能说明
==========

血压app顾问端的目的是为了方便顾问管理会员的健康，所有主要的功能都是围绕会员的健康信息，主要涉及下面的不同方面。

血压记录
-------

顾问可以在**血压记录**功能中查看其所有会员的血压统计数据，包括血压统计趋势图、系统生成的不同时间段的血压报告，根据这些血压数据，顾问可以很清楚的获得他的会员的健康状态，并根据这些数据（包括后面会提及的用药记录与体检报告），反馈会员合适的健康建议。

健康建议
-------

顾问可以查看自己曾经给某个会员提供的所有建议情况，包括审核与未审核的，其中审核过的建议指会员能够查看到的，未审核建议指，虽然顾问添加了此条建议，但是并不准备让会员看到。顾问根据自己所掌握的会员的健康数据，添加合理的健康建议反馈给会员，每次添加建议后，会员并不能直接看到，需要顾问确认无误后审核通过，会员才能够看到。

用药记录
-------

**用药记录**功能，顾问能够查看他的会员的用药情况，包括当前用药情况，历史用药情况，根据这些用药情况，顾问可以给用户更加精确的健康建议。

上传报告
-------

**上传报告**功能并不是指顾问可以上传报告，而是指顾问可以查看会员上传的所有报告，根据这些报告中的体征情况，顾问可以更加全面的了解他的会员，并给出精准建议。

会员个人信息
-----------

**会员个人信息**功能使顾问可以了解会员的信息，包括年龄，血型还有疾病史等很多重要的医学信息，这些信息将帮助顾问给出精准的健康建议。

沟通交流
-------

**沟通交流**使顾问与会员之间能够充分的交互，提高顾问的服务质量。

设置
----

**设置**包括顾问个人信息的修改，精确而详细的个人信息使对你感性的用户量大量增加。您也可以在这部分进行密码的修改；也可以了解app的相关内容。

活动
----

**活动**是血压管理系统对自己的宣传与推广所开展的一些优惠活动，在这一部分您将了解到系统曾经开展的和将要开展的活动信息，我们会在这些活动中宣传我们的理念和产品推广。
