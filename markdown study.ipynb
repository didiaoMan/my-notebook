# markdown 语法说明
[markdown官方网站说明](https://daringfireball.net/projects/markdown/syntax)

## 标题
1. 使用1-6个\#作为标记后跟空格，然后输入你的标题，从而建立1-6级的标题   

  例如：
  ```
  # 这是一个H1
  ## 这是一个H2
  ###### 这是一个H6
  ```
  效果如下：
  > # 这是一个H1
  > ## 这是一个H2
  > ###### 这是一个H6

2. 使用=和\-   
  例如:
  ```
  This is an H1
  =============
  
  This is an H2
  -------------
  ```
  效果如下:
  > This is an H1
  > =============
  > This is an H2
  > -------------

## 列表
### 无序列表
例如:
```
- 列表1
    - 列表1.1
    - 列表1.2
- 列表2
- 列表3
```
效果为:
> - 列表1
>     - 列表1.1
>     - 列表1.2
> - 列表2
> - 列表3

### 有序列表
例如:
```
1. 列表1
    1. 列表1.1
    2. 列表1.2
2. 列表2
3. 列表3
```
效果为：
> 1. 列表
     1. 列表1.1
     2. 列表1.2
2. 列表2
3. 列表3
  
## 引用
例如：
```
> This is a blockquote with two paragraphs. Lorem ipsum dolor sit amet,
> consectetuer adipiscing elit. Aliquam hendrerit mi posuere lectus.
```
效果为:
> This is a blockquote with two paragraphs. Lorem ipsum dolor sit amet,
> consectetuer adipiscing elit. Aliquam hendrerit mi posuere lectus.

## 粗体和斜体
例如：
```
*这是斜体*
**这是粗体**
```
效果为:
> *这是斜体*
> **这是粗体**

## 链接与图片
链接：格式\[显示文本\]\(链接地址\)   
图片: 格式!\[显示文本\]\(图片地址\)
例如：
```
[Github](https://github.com/didiaoMan)
![Logo](https://www.baidu.com/img/bd_logo1.png)
```
效果为：
> [Github](https://github.com/didiaoMan)
> ![Logo](https://timgsa.baidu.com/timg?image&quality=80&size=b9999_10000&sec=1508739665689&di=fcde43e38f7f8798acd972d0e6b46098&imgtype=0&src=http%3A%2F%2Fcdn-img.easylogo.cn%2Fgif%2F110%2F110849.gif)

## 分割线
例如：
```
文段1
***
文段2
```
效果：
>文段1
> ***
> 文段2

## 表格
例如：
```
| Item    | Value    | Qty |
| ----    | -----    | --- |
| Computer | 1600 USD  |  5  |
| Phone   |  12USD   | 12  |
```
效果为：
>| Item    | Value    | Qty   |
>| ----    | -----    | ---   |
>| Computer | 1600 USD  |  5    |
>| Phone   | 12USD    | 12    |
