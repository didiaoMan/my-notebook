血压app用户端功能说明
==========

---

血压管理app用户端，功能主要分为：
1. [血压管理](#bloodpressure_manage)
2. [用药记录](#medication)
3. [健康建议](#advice)
4. [上传报告](#upload)
5. [顾问](#consultant)
6. [活动](#activity)
7. [设置](#setting)

现分别对每个功能进行阐述。

<h2 id="bloodpressure_manage">1.血压管理</h2>

**血压管理**是app的主要功能，用户可以上传自己的血压数据到云端数据库，通过系统统一管理，系统会根据你所上传的数据，统计并为您生成个性的血压报告，您可以在app的**查看血压统计信息**功能中，查看自己的血压统计数据，包括：统计趋势折线图表、所有血压数据、以及每周或每两周或每月或全部数据所产生的血压报告。

<h2 id="medication">2.用药记录</h2>

**用药记录**也是app能够管理您的健康所必须的重要功能，在用药记录功能中，用户需要添加自己的当前用药记录，因为血压会随用药不同出现不同的波动，所有您的用药情况是您的顾问给你合理健康建议所必须的一个参考，在用药记录这个功能，用户不仅可以添加自己的当前用药情况，也可以添加曾经用药情况，并且所有的用药记录都会在云端保存，您的顾问将根据这些用药情况给您合理的健康建议，这也是我们系统所追求的目标。

<h2 id="advice">3.健康建议</h2>

**健康建议**是app对管理您的健康所做的一个反馈功能，在健康建议功能中，您将能够查看您的顾问提供给您的一些健康建议，包括注意事项、饮食推荐、或者是就医推荐等，您根据这些建议可以很好的改善并维持自己的健康状态。

<h2 id="upload">4.上传报告</h2>

**上传报告**使您能够上传您的体检报告，当你上传成功之后，您的顾问也将能够看到这些报告，并帮组分析您的体征，这也是顾问给您健康建议的重要参考。

<h2 id="consultant">5.顾问</h2>

**顾问**功能允许您选择您的顾问，并且在选择之前，您能查看所有顾问的详细信息；并且当您选定顾问之后，您可以与您的顾问进行沟通交流，详细询问您所关心的问题。

<h2 id="activity">6.活动</h2>

**活动**是app对管理系统所做的宣传，我们会开展一些活动，并将这些活动信息发布在这个功能模块，如果您对活动感兴趣，可以联系我们，通过这些活动我们来宣传我们产品的理念与目的。

<h2 id="setting">7.设置</h2>

在**设置**功能中，你可以修改你的个人信息，不过请认真填写个人信息，因为这些信息中有很多可以为您的顾问提供指导意义。并且您也可以在这个功能中修改您的密码，也可以查看app的介绍信息。
